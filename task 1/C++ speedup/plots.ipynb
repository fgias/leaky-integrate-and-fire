```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
u = pd.read_csv('u.txt', names=['u'])
N = len(u)
```

```python
plt.scatter(range(1,N+1), np.roll(u,500))
plt.ylim([0, 1])
plt.yticks([0, 0.5, 1])
plt.ylabel("ui")
plt.xticks([1, N/2, N])
plt.xlabel("index i")
plt.savefig('N=2000.pdf')
```
