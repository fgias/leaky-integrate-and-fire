```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
u = pd.read_csv('u.txt', names=['u'])
N = len(u)
```

```python
plt.scatter(range(1,N+1), u)
plt.ylim([0, 1])
plt.yticks([0, 0.5, 1])
plt.ylabel("ui")
plt.xticks([1, N/2, N])
plt.xlabel("index i")
plt.savefig('N=1000.pdf')
```
